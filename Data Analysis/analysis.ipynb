Handle Missing Values:

```python
#Missing Values in a Dataset means Not a Number
#if a column have more than 40% missing values then you can check - it is important or not , if it is important you can handle it and otherwise delete it
#if a column have less than 40% missing values then you can handle it

#categorical[mode] data means other than numeric column

avg_age = round(df['age'].mean())
print(avg_age)

```

```python
df['age'] = df['age'].fillna(avg_age)

df.isnull().sum()
```

```python
cols_to_drop = ['embarked']
df = df.drop(columns=cols_to_drop)
df.columns
```

```python
df = df.dropna(subset=['deck'])
df.isnull().sum()
```
